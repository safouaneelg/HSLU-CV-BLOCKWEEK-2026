# End-to-End Credit Risk Modeling: Handling Imbalanced Data

## 1. Project Overview

In financial lending, the cost of a "False Negative" (predicting a defaulter is safe) is significantly higher than a "False Positive" (predicting a safe borrower is risky). This notebook demonstrates a robust machine learning pipeline designed to predict the probability of a borrower experiencing serious financial distress in the next two years.

We will focus specifically on the challenges of **real-world "dirty" data**, including missing values, outliers, and severe class imbalance.

## 2. The Dataset: "Give Me Some Credit"

We use the popular dataset from the [Give Me Some Credit](https://www.kaggle.com/c/GiveMeSomeCredit) competition.

* **Target Variable:** `SeriousDlqin2yrs` (Binary: 0 for repaid, 1 for default).
* **Features:** Total of 10 features including `MonthlyIncome`, `DebtRatio`, and `NumberOfDependents`.
* **Key Challenge - Imbalance:** Roughly **93%** of the samples are negative (Class 0) and only **7%** are positive (Class 1). A naive model could predict "0" for everyone and achieve 93% accuracy, yet fail completely at its purpose.

## 3. Workflow & Methodology

To build a reliable model that avoids data leakage and handles imbalance, we will follow this pipeline:

1. **Stratified Splitting:** Ensuring the minority class is represented equally in Train and Test sets.
2. **Preprocessing Pipeline:**
   * **Imputation:** Handling missing values in `MonthlyIncome` and `Dependents` using Median strategies to avoid outlier skew.
   * **Scaling:** Using `RobustScaler` to normalize data while ignoring extreme financial outliers.
3. **Modeling:** Implementing a **Random Forest Classifier** with `class_weight='balanced'` to penalize misclassification of the minority class.
4. **Validation:** Using **Stratified K-Fold Cross-Validation** to rigorously test performance stability.
5. **Optimization:** performing a **GridSearchCV** to scientifically tune hyperparameters (`n_estimators`, `max_depth`, `min_samples_leaf`).
6. **Evaluation:** Moving beyond Accuracy to use **ROC-AUC**, **Recall**, and **Confusion Matrices**.

## 4. Tech Stack

* **Pandas & NumPy:** Data manipulation.
* **Scikit-Learn:** Model architecture, pipelines, and metrics.
* **Matplotlib & Seaborn:** Visualization of results.

## Step 1: Setup & Data Splitting

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# 1. Load Data
df = pd.read_csv('data/Give_Me_Some_Credit/cs-training.csv.zip')
df.head()
```

```python
df.isnull().sum()
```

### Add missing values

```python
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

# --- ARTIFICIALLY INTRODUCE MISSING VALUES ---
np.random.seed(42)

# Select only feature columns (exclude the Target 'SeriousDlqin2yrs')
# We rarely want to impute the Target variable; usually, we just drop those rows.
feature_cols = df.columns.drop('SeriousDlqin2yrs')

# Create a boolean mask where 2% of the cells are True
mask = np.random.rand(len(df), len(feature_cols)) < 0.02

# Insert NaNs where the mask is True
# We use .mask() which replaces values where condition is True
df[feature_cols] = df[feature_cols].mask(mask)

print(f"Artificial 'Dirty' Data Created.")
print(f"Total missing values in dataset: {df.isna().sum().sum()}")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.boxplot()
```

```python
# Define Target and Features
X = df.drop(columns=['SeriousDlqin2yrs'])
y = df['SeriousDlqin2yrs']

# 2. Stratified Train-Test Split (The "Holdout" Set)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% for final validation
    stratify=y,          # Crucial for imbalanced data!
    random_state=42
)

print(f"Training shapes: {X_train.shape}")
print(f"Default Rate in Train: {y_train.mean():.2%}")
print(f"Default Rate in Test:  {y_test.mean():.2%}")
```

## Step 2: Building the Preprocessing Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
```

```python
# Define which columns need what
# (In this dataset, all features are numeric, but this structure allows flexibility)
numeric_features = X.columns.tolist()

# Create the transformer block
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')), # Handle missing values
    ('scaler', RobustScaler())                     # Handle outliers
])

# Combine into a preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)
    ])
```

## Step 3: The Full Pipeline & Cross-Validation

```python
from sklearn.model_selection import StratifiedKFold, cross_validate
```

```python
# 3. Define the Full Pipeline (Same as before)
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(
        n_estimators=100, 
        random_state=42, 
        class_weight='balanced' 
    ))
])
```

```python
# 4. Stratified Cross-Validation with Multiple Metrics
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# We want to see all these metrics for every fold
scoring_metrics = ['accuracy', 'roc_auc', 'f1', 'recall']

cv_results = cross_validate(
    model_pipeline, 
    X_train, y_train, 
    cv=cv, 
    scoring=scoring_metrics
)

# Print the comparison
print(f"--- Cross-Validation Results (Mean of 5 Folds) ---")
print(f"Accuracy:  {cv_results['test_accuracy'].mean():.4f}  (Looks great, but misleading!)")
print(f"ROC-AUC:   {cv_results['test_roc_auc'].mean():.4f}  (The real performance)")
print(f"Recall:    {cv_results['test_recall'].mean():.4f}  (How many defaulters we caught)")
print(f"F1-Score:  {cv_results['test_f1'].mean():.4f}  (Balance of precision/recall)")
```

## Step 4: Final Evaluation

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, f1_score, recall_score, roc_auc_score
from sklearn.metrics import classification_report, confusion_matrix

# 5. Fit on full training data and Predict
model_pipeline.fit(X_train, y_train)
y_pred = model_pipeline.predict(X_test)
y_probs = model_pipeline.predict_proba(X_test)[:, 1]

# 6. The "Accuracy Paradox" Reveal
acc = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_probs)
f1 = f1_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)

print(f"--- Final Test Set Metrics ---")
print(f"1. Accuracy: {acc:.2%} (The Trap)")
print(f"   - If we predicted 'No Default' for everyone, we'd get ~93%.")
print(f"   - Our model gets {acc:.2%}, which seems similar, but look deeper:\n")

print(f"2. Evaluation Metrics (The Truth)")
print(f"   - ROC-AUC:  {auc:.4f} (Ability to rank risk)")
print(f"   - Recall:   {rec:.4f} (Percentage of defaulters caught)")
print(f"   - F1-Score: {f1:.4f} (Combined quality)")

print("\n--- Detailed Classification Report ---")
print(classification_report(y_test, y_pred))

# --- NEW: Confusion Matrix Visualization ---
print("\n--- Confusion Matrix ---")
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted: Safe (0)', 'Predicted: Default (1)'],
            yticklabels=['Actual: Safe (0)', 'Actual: Default (1)'])
plt.title('Confusion Matrix: Where are the errors?')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

**The model is ignoring the minority class.**

* **Recall (0.15):** You are catching only 15% of the actual defaulters.
* **Precision (0.55):** When the model *does* guess "Default," it's right about half the time.
* **The Verdict:** `class_weight='balanced'` helped slightly, but not enough. The signal from the majority class is still drowning out the defaulters.

### **The Decision: Upsampling vs. Downsampling**

For this specific dataset (Credit Risk), **I strongly recommend Upsampling.**

* **Why NOT Downsample?**
  * You have a 14:1 imbalance (\~28k Safe vs \~2k Default).
  * To downsample to a 1:1 ratio, you would have to throw away \~26,000 "Safe" borrowers.
  * **The Risk:** You are discarding 90%+ of your data. In credit risk, "Safe" borrowers are diverse. If you throw them away, the model loses the ability to recognize normal variations in safe behavior.
* **Why Upsample (specifically SMOTE)?**
  * You keep all the "Safe" data.
  * You artificially boost the "Default" data to match the size of the safe class.
  * **The Technique:** Instead of just duplicating rows (which causes overfitting), we use **SMOTE (Synthetic Minority Over-sampling Technique)**. It creates *new, synthetic* examples by interpolating between existing defaulters.

## Implementation: Adding SMOTE to the Pipeline

*Note: To put resampling inside a Pipeline, we need the `imbalanced-learn` library. It is the standard companion to scikit-learn for this task.*

**Install command (if needed):** `!pip install imbalanced-learn`

Here is the updated pipeline using `SMOTE`. Note that we switch to `imblearn.pipeline` because standard sklearn pipelines don't know how to handle resampling (they breaks during the `predict` phase if you don't use the imblearn wrapper).

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline 
```

```python
# 1. Define the Pipeline with SMOTE
# Notice the order: Imputer -> Scaler -> SMOTE -> Classifier
# SMOTE only runs during .fit(), never during .predict() (as intended)
resampling_pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),           # Prepare the data first
    ('smote', SMOTE(random_state=42)),        # Create synthetic defaulters
    ('classifier', RandomForestClassifier(
        n_estimators=100, 
        random_state=42
        # We don't need class_weight='balanced' anymore, SMOTE handles it
    ))
])
```

```python
# 2. Fit on Training Data
print("Training with SMOTE...")
resampling_pipeline.fit(X_train, y_train)
```

```python
# 3. Predict on Test Data
y_pred_smote = resampling_pipeline.predict(X_test)
```

```python
# 4. Compare Results
print("\n--- Classification Report (With SMOTE) ---")
print(classification_report(y_test, y_pred_smote))
```

```python
# 5. Visual Check
cm_smote = confusion_matrix(y_test, y_pred_smote)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_smote, annot=True, fmt='d', cmap='Greens', cbar=False) # Changed color to differentiate
plt.title('Confusion Matrix with SMOTE')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

Recall more than doubled (from 0.15 to 0.37). The model is no longer ignoring the minority class; it's actively trying to find them.
Precision dropped (from 0.55 to 0.37). Because the model is being more aggressive, it's making more "False Alarms."

Still long way to go..

## Grid Search with SMOTE Pipeline

```python
from sklearn.model_selection import GridSearchCV

# 1. Define the Parameter Grid
# We can tune steps inside the pipeline using 'stepname__parameter'
param_grid_smote = {
    # Tuning the Imputer (inside the preprocessor -> numeric_transformer)
    'preprocessor__num__imputer__strategy': ['mean', 'median'],
    
    # Tuning SMOTE (how many neighbors to use for generating samples)
    'smote__k_neighbors': [3, 5, 7],
    
    # Tuning Random Forest
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [10, 20, None],
    'classifier__min_samples_leaf': [1, 5] 
}
```

```python
# 2. Setup GridSearchCV
# Note: We must use the 'resampling_pipeline' defined earlier
grid_search_smote = GridSearchCV(
    resampling_pipeline,
    param_grid_smote,
    cv=3,                 # 3-fold CV to save time
    scoring='f1',         # Optimize for F1-Score (balance of Precision/Recall)
    n_jobs=-1,            # Use all cores
    verbose=1
)
```

```python
# 3. Fit the Grid Search
print("Starting Grid Search with SMOTE (this may take a minute)...")
grid_search_smote.fit(X_train, y_train)
```

```python
# 4. Best Results
print(f"\nBest Parameters: {grid_search_smote.best_params_}")
print(f"Best CV F1-Score: {grid_search_smote.best_score_:.4f}")
```

## Evaluation of Tuned SMOTE model

```python
# 1. Get Best Model
best_smote_model = grid_search_smote.best_estimator_

# 2. Predict on Test Set
y_pred_best = best_smote_model.predict(X_test)
y_probs_best = best_smote_model.predict_proba(X_test)[:, 1]

# 3. Final Report
print("\n--- Final Tuned SMOTE Classification Report ---")
print(classification_report(y_test, y_pred_best))
```

```python
# 4. Final Confusion Matrix
cm_best = confusion_matrix(y_test, y_pred_best)
plt.figure(figsize=(6, 5))
sns.heatmap(cm_best, annot=True, fmt='d', cmap='Purples', cbar=False)
plt.title('Final Confusion Matrix (Tuned SMOTE)')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
